```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import statistics as ss
```

```python
data=pd.read_csv(r"C:\Users\Abdelrahman\Downloads\Telegram Desktop\Titanic-Dataset2.csv")
data
duplicated=data.duplicated().sum()
print(f"sum of duplicated : {duplicated}")
```

```python
print(data.isnull().sum())
```

```python
median_age=data['Age'].median()
data['Age'].fillna(median_age , inplace=True)
print(f'Median Age used for missing Value : {median_age}')
```

```python
most_common_Embarked=data['Embarked'].mode()[0]
data['Embarked'].fillna(most_common_Embarked , inplace=True)
print(f"most_common_Embarked  value used : {most_common_Embarked}")
```

```python
data.drop(columns=['Cabin'],inplace=True)
```

```python
print (data.isnull().sum())
print(data.head())
```

```python
print(data.describe())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=data,x='Survived')
plt.title('Survival Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=data,x='Sex',y='Survived',color="red")
plt.title('Survival Rate by Sex')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=data,x='Pclass',y='Survived',color='skyblue')
plt.title('Survival Rate by Pclass')
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.histplot(data['Age'],kde=True,bins=30,color="skyblue")
plt.title('Age Distibution')

plt.subplot(1,2,2)
sns.histplot(data['Fare'],kde=True,bins=30,color="green")
plt.title('Fare Distibution')

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.histplot(data=data,kde=True,x='Fare',color="skyblue",hue='Survived',multiple='stack')
plt.title('Fasr and Survival')

plt.subplot(1,2,2)
sns.histplot(data=data,x='Age',hue='Survived',multiple='stack',kde=True)
plt.title('Age and Survival')

plt.show()
```
